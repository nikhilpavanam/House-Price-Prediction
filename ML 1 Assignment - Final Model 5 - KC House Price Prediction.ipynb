```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as stm
```

```python
house_train = pd.read_csv("wk3_kc_house_train_data.csv")
```

```python
house_train.head(10)
```

```python
house_train['age'] = 2020 - house_train.yr_built
```

```python
house_train['price_log'] = np.log(house_train.price)
```

```python
house_train['sqft_living_log'] = np.log(house_train.sqft_living)
```

```python
house_train['sqft_lot_log'] = np.log(house_train.sqft_lot)
```

```python
house_train['sqft_living15_log'] = np.log(house_train.sqft_living15)
```

```python
house_train['sqft_lot15_log'] = np.log(house_train.sqft_lot15)
```

```python
house_train['sqft_above_log'] = np.log(house_train.sqft_above)
```

```python
mask = house_train['sqft_basement'] >= 1
house_train.loc[mask, 'sqft_basement_bin'] = 1
```

```python
mask = house_train['sqft_basement'] < 1
house_train.loc[mask, 'sqft_basement_bin'] = 0
```

```python
zipcode_dummy = pd.get_dummies(house_train['zipcode'])
```

```python
zipcode_dummy
```

```python
house_train1 = pd.concat([house_train, zipcode_dummy], axis=1)
```

```python
house_train1.shape
```

```python
house_train1.head()
```

```python
house_train1.info()
```

```python
unwanted = ['id','date','price','sqft_living','sqft_lot','sqft_living15','sqft_lot15','sqft_living15_log','sqft_lot15_log','sqft_above','sqft_basement','yr_built','yr_renovated','zipcode','lat','long']
house_train1.drop(unwanted,axis=1,inplace=True)
```

```python
X = house_train1.drop(["price_log"], axis = 1)
y = house_train1['price_log']
```

```python
X.head(10)
```

```python
y.head(10)
```

```python
x = stm.add_constant(X)
x.head()
```

```python
mod5 = stm.OLS(y,x).fit()
mod5.params
```

```python
mod5.summary()
```
