```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as stm
```

```python
house_train = pd.read_csv("wk3_kc_house_train_data.csv")
```

```python
house_train.head(10)
```

```python
house_train['price_log'] = np.log(house_train.price)
```

```python
house_train.info()
```

```python
unwanted = ['id','date','price','yr_built','yr_renovated','zipcode','lat','long']
house_train.drop(unwanted,axis=1,inplace=True)
```

```python
X = house_train.drop(["price_log"], axis = 1)
y = house_train['price_log']
```

```python
X.head(10)
```

```python
y.head(10)
```

```python
x = stm.add_constant(X)
x.head()
```

```python
mod3 = stm.OLS(y,x).fit()
mod3.params
```

```python
mod3.summary()
```
